# EDA

Добавим в выборку sales_train колонки item_category_id и ID из выборок items и test соответсвенно.

```python
sales_train_new = pd.merge(sales_train, items.drop('item_name', axis=1), how='left', on='item_id')\
.merge(test, how='left', on=['shop_id','item_id'])
sales_train_new
```

Проверим полученную выборку на наличие NaN. Удалим появившиеся NaN. 

```python
sales_train_new.isnull().sum().sort_values(ascending=False)
```

```python
sales_train_new.dropna(inplace=True)
```

```python
# sales_train_new.fillna(0,inplace=True)
```

```python
sales_train_new
```

Столбец date разобьем на две части: месяц и год. Преобразуем новые признаки в числовой формат.

```python
sales_train_new.insert(2, "Month", sales_train_new.date.str.split(pat=".", expand=True).astype('int32')[1])
sales_train_new.insert(3, "Year", sales_train_new.date.str.split(pat=".", expand=True).astype('int32')[2])
```

Выделим целевой признак в отдельную выборку.

```python
y_train_full = sales_train_new.item_cnt_day
```

Удалим лишние столбцы.

```python
sales_train_new.drop(['date','item_cnt_day'], axis=1, inplace=True)
```

```python
# sales_train_new.drop(['date','date_block_num','item_cnt_day'], axis=1, inplace=True)
```

```python
# sales_train_new.drop(['date','date_block_num','item_cnt_day','item_price'], axis=1, inplace=True)
```

Выборка для обучения ML моделей примет вид:

```python
sales_train_new = sales_train_new.reindex(sorted(sales_train_new.columns), axis=1)
```

```python
sales_train_new
```

```python
test
```

Приведем тестовую выборку к виду тренировочной (добавим необходимые признаки):

```python
test.insert(0, "Month", 11)
test.insert(1, "Year", 2015)
test.insert(2, "date_block_num", 34)
test = test.merge(items.drop('item_name', axis=1), how='left', on='item_id')
```

```python
test = test.reindex(sorted(sales_train_new.columns), axis=1)
test
```

Признак item_price для тестовой выборки попробуем предсказать при помощи случайного леса:

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
search_price_train = sales_train_new.drop('item_price', axis=1)
price_train_full = sales_train_new.item_price

search_price_test = test.drop('item_price', axis=1)
```

Разбиение на тренировочную и валидационную выборки:

```python
sp_x_train, sp_x_val, price_train, price_val = train_test_split(search_price_train, price_train_full, test_size=0.2, random_state=1)
```

```python
rf_price = RandomForestRegressor(n_estimators=15,random_state=1, max_depth = 10)
rf_price.fit(sp_x_train, price_train)
```

```python
print(
    '''
    - Точность на тестовой выборке: {}
    - Точность на валидационной выборке: {}
    '''.format(
        rf_price.score(sp_x_train, price_train),
        rf_price.score(sp_x_val, price_val)
        )
)
```

```python
item_price_predict = rf_price.predict(search_price_test)
```

```python
item_price_predict.shape
```

```python
item_price_predict
```

Добавим в тестовую выборку найденный признак item_price, отсортируем колонки в порядке, аналогичном тренировочной выборке:

```python
test['item_price'] = item_price_predict
test = test.reindex(sorted(sales_train_new.columns), axis=1)
```

# ML модели

## Случайный лес

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
%%time

rf = RandomForestRegressor(n_estimators=15,random_state=1, max_depth = 20)
rf.fit(X_train, y_train)
```

```python
print(
    '''
    - Точность на тестовой выборке: {}
    - Точность на валидационной выборке: {}
    '''.format(
        rf.score(X_train, y_train),
        rf.score(X_val, y_val)
        )
)
```
